```python
pkts_features = [' Inbound', ' ACK Flag Count', ' Destination Port',
       'Init_Win_bytes_forward', ' Avg Fwd Segment Size', ' URG Flag Count',
       'Unnamed: 0.1', ' Max Packet Length', ' Fwd Packet Length Mean',
       ' Average Packet Size']
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import train_test_split
pkts = pkts.dropna()

X = pkts[pkts_features]
y = pkts['DataType']

X_train, val_X, y_train, val_y = train_test_split(X, y, test_size=0.20, random_state=1)


y_train = y_train.values.reshape(-1, 1)
y_train = y_train.ravel()
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier as MLPC
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier as DTC
from sklearn.ensemble import AdaBoostClassifier as ABC
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import mean_absolute_error
```

```python
pkts_model_Naive_Bayes = GaussianNB()
pkts_model_Naive_Bayes.fit(X_train, y_train)
pkts_preds_Naive_Bayes = pkts_model_Naive_Bayes.predict(val_X)
print(mean_absolute_error(val_y, pkts_preds_Naive_Bayes))
```

```python
pkts_model_ABC = ABC()
%time pkts_model_ABC.fit(X_train, y_train)
pkts_preds_ABC = pkts_model_ABC.predict(val_X)
%time pkts_score_ABC = pkts_model_ABC.score(val_X, val_y)
```

```python
pkts_model_DTC = DTC()
pkts_model_DTC.fit(X_train, y_train)
pkts_preds_DTC = pkts_model_DTC.predict(val_X)
print(mean_absolute_error(val_y, pkts_preds_DTC))
```

```python
pkts_model_MLPC = MLPC()
pkts_model_MLPC.fit(X_train, y_train)
pkts_preds_MLPC = pkts_model_MLPC.predict(val_X)
print(mean_absolute_error(val_y, pkts_preds_MLPC))
```

```python
pkts_model_SVC = SVC()
pkts_model_SVC.fit(X_train, y_train)
pkts_preds_SVC = pkts_model_SVC.predict(val_X)
print(mean_absolute_error(val_y, pkts_preds_SVC))
```

```python
pkts_model_RFC = RFC(random_state = 10)
pkts_model_RFC.fit(X_train, y_train)
pkts_preds_RFC = pkts_model_RFC.predict(val_X)
print(mean_absolute_error(val_y, pkts_preds_RFC))
```
